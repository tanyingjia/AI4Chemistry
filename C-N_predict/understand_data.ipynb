```python
from sklearn.model_selection import train_test_split
# data = pd.read_csv('./data/data.csv')
train_data, test_data = train_test_split(data,test_size=0.3, random_state=2025)
```

```python
test_data.iloc[:,5:]
```

```python
from sklearn.ensemble import RandomForestRegressor   #导入必要的库，模型
from sklearn.metrics import root_mean_squared_error  # 导入评估方法 rmse
RF = RandomForestRegressor()  #实例化模型
RF.fit(train_data.iloc[:,5:], train_data.iloc[:, 4]) #训练模型
test = RF.predict(test_data.iloc[:,5:]) # 预测，模型
print('模型在测试集上的rmse：', root_mean_squared_error(test_data.iloc[:, 4], test))  #在测试集上评估模型
```

```python
from sklearn.model_selection import KFold
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error

NFOLD = 5
folds = KFold(n_splits=5,shuffle=True, random_state=5)
val_set = [] 
rmse_set = []
test_set = []
for k, (train_idx, val_idx) in enumerate(folds.split(train_data)):
    train_X, val_X = train_data.iloc[train_idx, 5:], train_data.iloc[val_idx, 5:]
    train_y, val_y = train_data.iloc[train_idx, 4], train_data.iloc[val_idx, 4]

    rf = RandomForestRegressor()
    rf.fit(train_X, train_y)
    val_yield = rf.predict(val_X)
    val_rmse = root_mean_squared_error(val_y, val_yield)
    print(f'第 {k+1} 折 验证集 RMSE：', val_rmse)

    val_set.append(val_yield)
    rmse_set.append(val_rmse)

    test_yield = rf.predict(test_data.iloc[:,5:])
    test_set.append(test_yield)
test_mean_yield = np.mean(test_set, axis=0)
test_rmse = root_mean_squared_error(test_data.iloc[:,4],test_mean_yield)
print('测试集的RMSE：', test_rmse)
```
